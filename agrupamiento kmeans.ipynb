```python
#importar las librerías para los gráficos
import matplotlib.pyplot as plt
import seaborn as sns
#esta linea se usa para no tener que usar plt.show()

%matplotlib inline
```

```python
# importar numpy python
import numpy as np
```

```python
# importar el sample generator de scikit-learn - make blobs
from sklearn.datasets.samples_generator import make_blobs
```

```python
X,Y = make_blobs(n_samples=1000, centers=5)
plt.scatter(X[:,0],X[:,1],s=50)
```

```python
# K-means de scikit-learn
from sklearn.cluster import KMeans
```

```python
# K-means
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(X)
```

```python
y_means = kmeans.predict(X)
```

```python
centers = kmeans.cluster_centers_
print(centers)
```

```python
plt.scatter(X[:,0],X[:,1],c=y_means,cmap='viridis')
plt.scatter(centers[:,0],centers[:,1],c='black',s=200,alpha=0.5)
```
